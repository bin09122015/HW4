```python
import nltk

shakespeare = open("t8.shakespeare.txt", "r")
text = shakespeare.read()

print (len(text))

# nltk.tokenize will take a long time to run
words = nltk.tokenize.word_tokenize(text)

print (len(words))
```

```python
nltk_text = nltk.Text(words)

nltk_text.concordance("creature")
```

```python
nltk_text.similar('beauty')
```

```python
nltk_text.common_contexts(['creature', 'beauty'])
```

```python
from nltk.draw.dispersion import dispersion_plot
# plot will be displayed in a separate window
dispersion_plot(nltk_text, ["creature", "beauty", "freedom", "war", "time", "memory"])
```

```python
from nltk import FreqDist
freq_dist = FreqDist(nltk_text)
freq_dist
```

```python
vocabulary = list(freq_dist.keys()) # list of all the distinct types in the text
vocabulary[:3] # look at first 3
```

```python
words = set(nltk_text)
long_words = [word for word in words if len(word) > 15]
sorted(long_words)
```
